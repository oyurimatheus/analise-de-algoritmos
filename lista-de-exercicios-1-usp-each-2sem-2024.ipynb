#### 1) Usando as definições das notações assintóticas $\Theta$, $O$ e $\Omega$, mostre que:

a) $1000n^2 = O(n^2)$
Ou seja, 

$1000n^2 \leq c*n^2$ $\forall c \gt 0;n_0 \gt 0$

Se $c > 1000$, logo a condição se satisfaz e temos que $1000n^2 = O(n^2)$

-----------

b) $5n^3 + 1000n^2 = O(n^4)$
Ou seja, 

$5n^3 + 1000n^2 \leq c*n^4$ $\forall c \gt 0;n_0 \gt 0$

Ao dividirmos ambos os lados por $n^2$, temos
$5n + 1000 \leq c*n^2$

Para $n$ suficientemente grande, $nˆ2$ sempre será maior que $5n$. Logo, a condição se satisfaz. 

-----------

c) $\frac{n^2}{1000} = \Omega(n^2)$
Ou seja,
$\frac{n^2}{1000} \geq c * n^2$ $\forall c \gt 0;n_0 \gt 0$

Dividindo ambos os lados por $n^2$:

$\frac{1}{1000} \geq c$

Para valores de $c \leq \frac{1}{1000}$, podemos escolher $c = \frac{1}{1000}$ e $n_0 = 1$, para todo $n \geq n_0$

Podemos aplicar na definição:

$$\frac{n^2}{1000} \geq c * n^2$$

Isto é:

$$\frac{n^2}{1000} \geq \frac{1}{1000} * n^2$$

Isso vale para todo $n \gt 1$ $\therefore$ a definição é válida

-----------

d) $n^4 - 25n^2 = \Omega(n^3)$
Ou seja,
$n^4 - 25n^2 \geq c * n^3$ $\forall c \gt 0;n_0 \gt 0$

Ao dividir ambos os lados por $n^2$:

$n^2 - 25 \geq c * n$

Aqui já conseguimos pensar que para um $n$ suficientemente grande, $n^2$ sempre será maior que uma constante multiplicado por $n$. Contudo, podemos fazer mais algumas transformações

$$n^2 - 25 \geq c * n$$

$n^2 - 25 - c * n \geq 0$

Isso torna uma inequação do segundo grau:
$\to$ $n^2 - c * n - 25 \geq 0$

Podemos aplicar a fórmula da equação do segundo grau:

$$
n = \frac{- (-c) \pm \sqrt{c^2 - 4 * 1 * (-25)}}{2 * 1}
$$

Resolvendo a equação, chegamos em $n_1 \cong 5$ e $n_2 \cong -4$. Discartamos a raiz negativa, já que estamos interessados apenas no valor positivo. 

Portanto, para $n \geq 6$ a condição se mantém válida.

-----------

e) $2^16n^2 \neq \Omega(n^3)$

Podemos falar que, 
$2^{16}n^2 \geq c * n^3$ $\forall c \gt 0;n_0 \gt 0$ **deve ser falso**

Simplificando ambos os lados por $n^2$:
$2^{16} \geq c * n$

Para $n$ suficientemente grande, o lado direito ($c * n$) sempre será maior que $2^{16}$.

Por exemplo, para $c = 2$ e $n = 2^{16}$ teríamos

$2^{16} \geq 2 * 2^{16}$

O que não é verdade. Portanto, $2^{16}n^2 \neq \Omega(n^3)$

-----------

f) $n^3 - 10n^2 \neq O(n^2)$

$n^3 - 10n^2 \leq c * n^2$ $\forall c \gt 0;n_0 \gt 0$ **deve ser falso**

Simplificando por $n^2$

$n - 10 \leq c$

Para $n$ suficientemente grande, o lado esquerdo ($n - 10$) sempre será maior que $c$.

Podemos pegar um valor como $c = 1$ e $n = 12$. Dessa forma, 

$$n - 10 \leq c$$

$$12 - 10 \leq 1$$

Isso nos dá: $2 \leq 1$, o que é **falso**

-----------

g) $4n^3 - 300n^2 + 7000n = \Theta(n^3)$
Ou seja, 
$c_1 * n^3 \leq 4n^3 - 300n^2 + 7000n \leq c_2 * n^3$ $\forall c_1 \gt 0; c_2 \gt 0;n_0 \gt 0$ 

Podemos simplificar ambos os lados por $n^3$:

$c_1 \leq 4 - \frac{300}{n} + \frac{7000^2}{n^2} \leq c_2$


Quando falamos da notação $\Theta$ estamos querendo dizer que uma função é limitada superior e inferiormente por $g(n)$ 

Podemos analisar os limites da inequação. 

Pegando o limite superior: 
$4 - \frac{300}{n} + \frac{7000^2}{n^2} \leq c_2$

Quando $n \to \inf$, o lado esquedo $4 - \frac{300}{n} + \frac{7000^2}{n^2}$ tende a $4$

Ou seja, para um valor de $c_2 \geq 4$ a condição se mantém válida

Já no limite inferior: 
$c_1 \leq 4 - \frac{300}{n} + \frac{7000^2}{n^2}$

Quando $n \to \inf$, o lado direito $4 - \frac{300}{n} + \frac{7000^2}{n^2}$ tende a $4$

Ou seja, para um valor de $c_1 \leq 4$ a condição se mantém válida

Então para $c_1 \leq 4$, $c_2 \geq 4$ e quando $n \to \infty$ a condição se satisfaz


-----------

h) $n^2 + nlgn = \Theta(n^2)$

Ou seja, 
$c_1 * n^2 \leq n^2 + nlgn \leq c_2 * n^2$ $\forall c_1 \gt 0; c_2 \gt 0;n_0 \gt 0$ 

Podemos simplificar ambos os lados por $n$:

$c_1 * n \leq n + lgn \leq c_2 * n$


Quando falamos da notação $\Theta$ estamos querendo dizer que uma função é limitada superior e inferiormente por $g(n)$ 

Podemos analisar os limites da inequação. 

Pegando o limite superior: 
$n + lgn \leq c_2 * n$

Isto é:

$lgn \leq c_2 * n - n$

Deixando o $n$ em evidência:

$lgn \leq n * (c_2 - 1)$

Quando $c_2 \gt 1$ a condição é verdadeira

Já no limite inferior: 

$c_1 * n \leq n + lgn$

Isto é, 

$c_1 * n - n \leq lgn$

Deixando o $n$ em evidência:

$n * (c_1 - 1) \leq lgn$

Quando $c_1 = 1$, a condição se satisfaz, portando podemos dizer que a condição é verdadeira

#### 4) Escreva uma função iterativa que recebe um vetor a de valores inteiros e um valor x, e determina todos os pares (i, j) de índices tais que $a[i] + a[j] = x$

-----------

```python
def encontra_soma(vetor, x):
    pares = []
    for i in range(len(vetor)):
        for j in range(1 + i, len(vetor)):
            if vetor[i] + vetor[j] == x:
                pares.append((i, j))
    
    
    return pares
                
print(f'vetor: [1, 2, 3, 2, 5, 0, 4], x = 4 ->pares: {encontra_soma([1, 2, 3, 2, 5, 0, 4], 4)} ')
```

#### 5) Escreva uma função iterativa que recebe duas matrizes `A` (de dimensão $n × m$) e `B` (de dimensão $m × p$) de valores inteiros, e devolve o produto de `A` por `B`.

-----------

```python
def multiplicacao_matrizes(A, B):
    
    linhas_a = len(A)
    colunas_a = len(A[0])

    colunas_b = len(B[0])
    
    matriz_resultante = [[0 for _ in range(colunas_b)] for _ in range(linhas_a)]

    for i in range(linhas_a):
        for j in range(colunas_b):
            for k in range(colunas_a):
                matriz_resultante[i][j] += A[i][k] * B[k][j]

    return matriz_resultante
    

matriz_A = [ # 3x2
    [2, 2],
    [3, 3],
    [4, 4],
]

matriz_B = [ # 2x3
    [5, 5, 5],
    [5, 5, 5],
]

print(multiplicacao_matrizes(matriz_A, matriz_B))
```

#### 6) Escreva uma função recursiva que recebe um valor inteiro `n` e devolve seu fatorial.

-----------

```python
def fatorial(n):
    if n < 1:
        return 1
    
    return n * fatorial(n - 1)


print(f'5! = {fatorial(5)}')
print(f'4! = {fatorial(4)}')
print(f'3! = {fatorial(3)}')
print(f'2! = {fatorial(2)}')
print(f'1! = {fatorial(1)}')
print(f'0! = {fatorial(0)}')
```

#### 7) Escreva uma função recursiva que calcula o termo $a_i$ de uma progressão aritmética de termo inicial $a_0$ e razão $r$ (obs: a função a ser implementada não será a forma mais eficiente de determinar o valor do termo $a_i$, mas a ideia aqui é exercitar o pensamento recursivo).

-----------

```python
def progressao_aritmetica(termo_buscado, termo_inicial, razao):
    if termo_buscado == 0:
        return termo_inicial
    
    return razao + progressao_aritmetica(termo_buscado - 1, termo_inicial, razao)

print(progressao_aritmetica(5, 1, 1))
print(progressao_aritmetica(9, 0, 2))
print(progressao_aritmetica(1, 0, 2))
```

#### 8) Escreva uma função recursiva que recebe dois valores inteiros $c$ e $n$, e devolve o valor de $c^n$ (sem recorrer ao uso de funções prontas que implementam a exponenciação).

-----------

```python
def potencia(c, n):
    if n == 0:
        return 1
    
    return c * potencia(c, n - 1)


print(potencia(2, 2))
print(potencia(2, 3))
print(potencia(3, 3))
print(potencia(2, 10))
```
